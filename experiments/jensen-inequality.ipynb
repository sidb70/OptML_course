```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**2

X = np.linspace(-10, 10, 100)
y = f(X)
plt.plot(X, y)
plt.show()
```

```python
m = 20
pts = (np.random.rand(m, 1) * 20 - 10).flatten()

for i in range(m):
    plt.plot(pts[i], f(pts[i]), 'ro')
plt.plot(X, y)
plt.show()
```

```python
num_trials = 100
for i in range(num_trials):
    lambda_ = np.random.rand(m,1).flatten()
    lambda_ /= lambda_.sum()

    lhs = f(lambda_.dot(pts))
    rhs = lambda_.dot(f(pts))

    print(lhs, rhs)

    assert lhs <= rhs
```
